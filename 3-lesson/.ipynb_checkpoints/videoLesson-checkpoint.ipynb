```python
from sklearn.metrics import mean_squared_error

feats = ['LSTAT','PTRATIO','TAX','RAD','DIS','RM','NOX','ZN','CRIM']

def create_model (X_train, Y_train, X_test, Y_test, feats, model):
    model.fit(X_train.loc[:,feats],Y_train)
    Y_pred=model.predict(X_test.loc[:,feats])
    mse = mean_squared_error(Y_test,Y_pred)
    print(mse)
    return mse

mse = create_model(X_train_scaled,Y_train,X_test_scaled,Y_test,feats, LinearRegression())
print (mse)
```

# Логистическая регрессия

Логистическая регрессия используется для задач классификации, а не задач прогнозирования значений признаков. Логистическая регрессия. Суть логистической регресси это отображение результатов линейной регрессии через сигмоидную функцию для вычисления нормированной величины от 0 до 1. Принято считать это вероятностной мерой для данной регрессии. Используется в бинарной классификации для двух классов.
z = w0+w1x1+...+wNxN
f(x)= 1/(1+exp(-z))
f(x) имеет область значений [0,1]
f(x) для класса 1
1-f(x) для класса 2


LogLoss 

Логистическая регрессия работает с lobLoss
logloss = -Fln(f)-(1-F)ln(1-f)
F- истинное значение
f=f(x) значение логистической регрессии
При идеальных ответах logLoss равен нулю. Максимальное значение не нормировано. Но чем оно больше -тем хуже

```python
df = pd.read_csv('train.csv')
print(df.head(10))
df = df.set_index("PassengerId")

targetProp = 'Survived'
y = df[targetProp]
x = df.drop(targetProp,axis=1)
x.info()
```

```python
# droping cabin, name, ticket. Because these fields are not fullfield completely
x = x.drop(['Cabin','Name','Ticket'],axis=1)
# adding means to the age
mean_age = x['Age'].mean()
print(x)
x['Age'] = x['Age'].fillna(mean_age)
x['Age'].unique()
```

```python
# embarked filling
x['Embarked'].value_counts()
```

```python
embarked_mode = x['Embarked'].mode()[0]
x['Embarked'] = x['Embarked'].fillna(embarked_mode)
x['Embarked'].value_counts()
```

```python
x.info()
```

```python
# filling sex as an integr values:
x['Sex']=(x['Sex']=='female').astype(int)
# making dummy variables for embarked
# dummies mean separating the single column to subcolumns.
# Subcolumn number is equal to the number of categories of original property
x = pd.get_dummies(x)
x.columns
```

```python
x.info()
```

```python
# All indicator parameters have to be transformed to the dummies
x['Pclass'] = x['Pclass'].astype('category')
x = pd.get_dummies(x)
x.columns
```

```python
from sklearn.linear_model import LogisticRegression

# Dividing the original data to train and validation data
# lets make the simplest model and comparing their quality with the more sophisticate algorithm
x_train, x_valid, y_train, y_valid = train_test_split(x,y,test_size=0.25,random_state=42)

log_reg = LogisticRegression(max_iter=10000)
log_reg.fit(x_train,y_train)
y_pred = log_reg.predict(x_valid)
y_pred
```

```python
# applying metrics
from sklearn.metrics import accuracy_score

print ("ac="+str(accuracy_score(y_valid,y_pred)))

y_probability = log_reg.predict_proba(x_valid)
print (y_probability)
# All probabilities have to be rounded to 0 or 1
# The first column is about the 0 class. The second is about the 1 class

#saving
x_train.to_pickle('x_train.pkl')
y_train.to_pickle('y_train.pkl')
x_valid.to_pickle('x_valid.pkl')
y_valid.to_pickle('y_valid.pkl')
```

# Метод опорных векторов

Метод для классификации и регрессии. Для классификации используется пространство признаков. Алгоритм работает по принципу максимизации дистанции между точка пространства признаков

```python
from sklearn.svm import SVC
from sklearn.preprocessing import MinMaxScaler

# Support vector machine is sensible for non scaled data. That is why
# we have to use scaler
scaler = MinMaxScaler()

cols_for_scale = ['Age', 'SibSp', 'Parch', 'Fare']
x_train[cols_for_scale] = scaler.fit_transform(x_train[cols_for_scale])
print(x_train)

x_valid[cols_for_scale] = scaler.transform(x_valid[cols_for_scale])
print(x_valid)
x_valid.describe()
```

```python
# persisting the scaler to the file
#from sklearn.externals import joblib
import joblib
joblib.dump(scaler, 'min_max_scaler.pkl')
scaler = joblib.load('min_max_scaler.pkl')
```

```python
# Classifying:--------------------------------------
svc = SVC()
svc.fit(x_train,y_train)
y_pred=svc.predict(x_valid)
y_pred_train = svc.predict(x_train)

# accuracy:-----------------------------------------
ac = accuracy_score(y_valid,y_pred)
print ('Accuracy 1:'+str(ac))
ac = accuracy_score(y_train,y_pred_train)
print ('Accuracy 2:'+str(ac))


```

```python
# Enhanced SVC:-------------------------------------
c_vals = np.logspace(-4,-1,100) # [10^-2,10^5]
print(c_vals)

accuracy_valid = []
accuracy_train = []
for val in c_vals:
    clf=SVC(C=val)
    clf.fit(x_train,y_train)
    y_pred = clf.predict(x_valid)
    y_pred_train = clf.predict(x_train)
    acc_valid = accuracy_score(y_valid,y_pred)
    acc_train = accuracy_score(y_train,y_pred_train)
    accuracy_valid.append(acc_valid)
    accuracy_train.append(acc_train)
    print('C={} \n\t acc_valid={} \n\t acc_train={}\n'.format(val,acc_valid,acc_train))


# 30 min
```

```python
import matplotlib.pyplot as plt
%config InlineBackend.figure_format = 'svg'
%matplotlib inline

plt.plot(c_vals,accuracy_valid)
plt.plot(c_vals,accuracy_train)
plt.xlabel('Parameter С')
plt.ylabel('Accuracy')
plt.legend(['valid','train'])
```

# Алгоритм KNN

```python
import pandas as pd
import numpy as np
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
from pylab import rcParams
%config InlineBackend.figure_format='svg'
%matplotlib inline

from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import RobustScaler

x_train = pd.read_pickle('x_train.pkl')
y_train = pd.read_pickle('y_train.pkl')
x_valid = pd.read_pickle('x_valid.pkl')
y_valid = pd.read_pickle('y_valid.pkl')

# Robust scaler is applied instead of min/max scaler
# Масштабирование признаков с минимизацией выбросов засчет использования медианы и расстояния между первой и третьей квартилью 
cols_for_scale = ['Age','SibSp','Parch','Fare']
scaler = RobustScaler()

x_train[cols_for_scale] = scaler.fit_transform(x_train[cols_for_scale])
x_train.describe()
```

```python
neighbors_number_array = np.arange(2,10)
accuracy_valid = []
accuracy_train = []

for neighbors_number in neighbors_number_array:
    clf = KNeighborsClassifier(n_neighbors=neighbors_number)
    clf.fit(x_train,y_train)
    y_pred = clf.predict(x_valid)
    y_pred_train = clf.predict(x_train)
    acc_valid = accuracy_score(y_valid,y_pred)
    acc_train = accuracy_score(y_train,y_pred_train)
    accuracy_valid.append(acc_valid)
    accuracy_train.append(acc_train)
    print('n_neighbors = {} \n\t acc_valid={} \n\t acc_train = {}\n'.format(neighbors_number,acc_valid,acc_train))
    
```

```python
accuracy_valid
```

```python
plt.plot(neighbors_number_array,accuracy_valid)
plt.plot(neighbors_number_array,accuracy_train)
plt.xlabel('n_neighbors')
plt.ylabel('accuracy')
plt.legend(['valid','train'])
```

# Деревья

Деревья решений относятся к наиболее интерпретируемым решениям. Так же не важна регуляризация и наличие линейных зависимостей. Для решения задач регрессии применяется модуль под названием DecisionTreeRegressor
Для решения задач классификации применяется модуль под названием DecisionTreeClassifier

```python
from sklearn.tree import DecisionTreeClassifier

x_train = pd.read_pickle('x_train.pkl')
y_train = pd.read_pickle('y_train.pkl')
x_valid = pd.read_pickle('x_valid.pkl')
y_valid = pd.read_pickle('y_valid.pkl')

# Для деревьев не нужно вводить нормировки значений - главное это выбрать нужные точки разбиения дерева
max_depth_arr = np.arange(2,20)

accuracy_valid = []
accuracy_train = []
for depth in max_depth_arr:
    # random state=>контролируемое перемешивание:
    clf=DecisionTreeClassifier(max_depth=depth,random_state=100)
    clf.fit(x_train,y_train)
    y_pred=clf.predict(x_valid)
    y_pred_train = clf.predict(x_train)
    acc_valid = accuracy_score(y_valid,y_pred)
    acc_train = accuracy_score(y_train,y_pred_train)
    accuracy_valid.append(acc_valid)
    accuracy_train.append(acc_train)
    print('max_depth = {} \n\t acc_valid={} \n\t acc_train={}\n'.format(depth,acc_valid,acc_train))
```

```python
plt.plot(max_depth_arr,accuracy_valid)
plt.plot(max_depth_arr,accuracy_train)
plt.xlabel('max_depth')
plt.ylabel('accuracy')
plt.legend(['valid','train'])
```

При увеличении глубины дерева точность растет на тренировочном наборе данных. При этом точность снижается на валидационном наборе данных
