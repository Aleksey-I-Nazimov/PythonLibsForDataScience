```python
import numpy as np
import pandas as pd

from scipy.stats import mode
import warnings
warnings.filterwarnings('ignore')

DATASET_PATH = './creditcard.csv'

data_frame = pd.read_csv(DATASET_PATH, sep=',')
print("DataFrame log "+str(data_frame.head()))

def select_color (value):
    if value==1:
        return 'red'
    else:
        return 'gray'

legend_dict = {'family':'Calibri','size':10}

unique_class_seriese = data_frame['Class'].unique()
print("Unique class series: "+str(unique_class_seriese))
for selected_class in unique_class_seriese:
    print("Analyzing: "+str(selected_class))
    selected_data_frame = data_frame[data_frame['Class']==selected_class]
    unique_values = np.asarray(selected_data_frame.value_counts())
    print(unique_values)
    # Normal scale:---------------------------------------------------------
    plt.hist(unique_values,color=select_color(selected_class),edgecolor='black',alpha=0.5,bins=20,orientation='horizontal',ec='white',log=False)
    plt.legend(labels=[str(selected_class)],loc='upper right',prop=legend_dict,frameon=False)
    plt.show()
    
    # Log scale:------------------------------------------------------------
    plt.hist(unique_values,color=select_color(selected_class),edgecolor='black',alpha=0.5,bins=20,orientation='horizontal',ec='white',log=True)
    plt.legend(labels=[str(selected_class)],loc='upper right',prop=legend_dict,frameon=False)
    plt.show()
    
    
```
