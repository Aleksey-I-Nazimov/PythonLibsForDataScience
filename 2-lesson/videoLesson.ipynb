```python
planets = ["Mercury","Venus","Earth","Mars","Jupiter","Saturn","Uranus","Neptune"]
masses = [0.0552,0.815,1.0,0.107,317.8,95,14.6,17.147]
plt.plot(planets,masses)
plt.show()
```

```python
plt.plot(planets,masses)
plt.yscale(value='log')
plt.show()
```

```python
planet_info = pd.Series(masses,index=planets)
planet_info
```

```python
planet_info.plot(kind='bar',logy=True)
plt.show()
```

```python
planet_info.plot(kind="barh",logx=True)
plt.show()
```

```python
url = 'https://infographics.economist.com/2018/databank/BMFile2000toJan2018.xls'
bmi = pd.read_excel(url)
print(bmi.shape)
print(bmi.columns)
bmi.iloc[:, :4].head(10)
```

```python
#Extracting the new data indexed by country and getting the series marked as the dollarPrice:-----------------
from pylab import rcParams
rcParams['figure.figsize'] = 9,12

bm_price_series = bmi.set_index('Country')['dollar_price']
print(bm_price_series.head())
bm_price_series.sort_values().plot(kind='barh')
plt.show()

bm_price_series.sort_index(ascending=False).plot(kind='barh')
plt.show()
```

```python
rcParams['figure.figsize']=4.5,6
plt.hist(bm_price_series,ec='white',color='lightgrey')
plt.show()
```

```python
plt.hist(bm_price_series,ec='white',color='lightgrey')
plt.savefig('bm_price',fmt='png')
plt.show()
```

```python
x = np.linspace(-5,5,101)
print(x)
x2 = x*x
print(x2)

x= np.linspace(-5,5,101)
print(x)
x3 = x*x*x
print(x3)


%config InlineBackend.figure_format='svg'
title_dict = {'fontsize':20,'fontweight':'bold','color':'#808080','family':'Calibri'}
legend_dict = {'family':'Calibri','size':10}

#plt.plot(x,x2,label='X2',color='#e3bb08f',linestyle='dashed')
#plt.plot(x,x3,label='X3',color='#e97451',linestyle='dotted')
plt.plot(x,x2,label='X2',linestyle='dashed')
plt.plot(x,x3,label='X3',linestyle='dotted')

plt.axis([-5,5,-8,8]) # horizontal and vertical legend
plt.xlabel("x",fontdict=title_dict)
plt.ylabel("y",fontdict=title_dict)

legend=plt.legend(labels=['x2','x3'],loc='lower right',prop=legend_dict,frameon=False)
plt.setp(legend.get_texts(),color='grey')
plt.title("Parabola", fontdict=title_dict,loc='left')

plt.grid(axis='y',color='lightgrey')

plt.show()
```

Figure  axes -    Matplotlib.  axes   figure

```python
fig,ax = plt.subplots(nrows=2,ncols=2)
ax1,ax2,ax3,ax4 = ax.flatten()
ax1.plot(x,x2)
ax1.set_xlim([-3,3])
ax4.plot(x,x3)
fig.set_size_inches(8.2,3.4)
plt.subplots_adjust(wspace=0.9,hspace=0.3)
```
